# Chemical simulation interface
This notebook provides an interface for a toy model of a chemical reaction in the gas phase.

## Import libraries

```python
%load_ext autoreload
%autoreload 2

%matplotlib widget 

import ipywidgets as widgets
from matplotlib import pyplot as plt
    
from collider import Simulation
from ipycanvas import Canvas
```

## Initialize and run the simulation
Initialize the Simulation class with the desired parameters and run the simulation loop.

```python
simulation = Simulation(Canvas, 800, 600, 200, temperature=500)
#simulation.canvas
```

```python
#simulation.run()
```

```python
with plt.ioff():
    fig = plt.figure(figsize=(8, 4))
    
widgets.VBox([simulation.canvas, fig.canvas], 
             layout=widgets.Layout(width='800px'))
```

```python
simulation.run(fig=fig)
```
